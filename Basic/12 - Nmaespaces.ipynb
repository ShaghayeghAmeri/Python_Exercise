### locals, Inclosing

```python
x = 'global'
def f():
    x = 'local f(), inclosing g()'
    print(locals())
    
    def g():
        x = 'local g()'
        print(locals())
        print(x)
        
    g()
```

```python
f()
```

```python
type(globals())
```

Locals return copy. Globals return refrence.

```python
def f():
    x = 2
    locals()['x'] = 999
    print(x)
```

```python
f()
```

```python
x = 2
globals()['x'] = 999
print(x)
```

### The global Decleration

```python
x = 3
def f():
    global x
    
    x = 999
    print(x)
```

```python
f()
```

```python
x
```

```python
def f():
    x = 2
    
    def g():
        nonlocal x
        x = 3
        
    g()
    print(x)
```

```python
f()
```
